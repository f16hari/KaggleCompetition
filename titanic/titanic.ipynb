```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
# Dataset preparation
dataset = pd.read_csv("train.csv")

features = ["Age", "Pclass", "Sex", "SibSp", "Parch", "Fare", "Embarked"]
label = "Survived"
categorical_features = ["Sex", "Embarked"]

dataset = dataset.dropna(subset=features)

Y = dataset[label].copy()
X = dataset[features].copy()

label_encoders = {}
for cf in categorical_features:
    label_encoder = LabelEncoder()
    label_encoder.fit(X[cf])
    label_encoders[cf] = label_encoder

for cf, label_encoder in label_encoders.items():
    X[cf] = label_encoder.transform(X[cf])

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8)

X
```

```python
model2 = svm.SVC()
result = model2.fit(x_train, y_train)
print(result.score(x_train, y_train))
print(result.score(x_test, y_test))

#final training model
model = svm.SVC()
result = model.fit(X, Y)
print(result.score(X, Y))
```

```python
# Testing on given competition training 
testDataSet = pd.read_csv("test.csv")
testDataSet[features] = testDataSet[features].fillna(0)
testDataSet = testDataSet.reset_index()

test_X = testDataSet[features].copy()

for cf, label_encoder in label_encoders.items():
    test_X[cf] = label_encoder.transform(test_X[cf])

predictions = model.predict(test_X)
final_output = pd.DataFrame(columns=["PassengerId", label])
final_output["PassengerId"] = testDataSet.loc[:, "PassengerId"]
final_output[label] = predictions

final_output.to_csv("svm_output.csv", index=False)
```
